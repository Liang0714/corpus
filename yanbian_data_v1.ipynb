```python
def regroup_year_catalogue(year_catalogue):
    catalogues = {}
    for index, r in year_catalogue.iterrows():
        year = index[0]
        cat = index[1]
        mean = r['mean']
        if cat not in catalogues:
            catalogues[cat] = {'year':[], 'mean':[]}
        catalogues[cat]['year'].append(year)
        catalogues[cat]['mean'].append(mean)
    return catalogues
```

```python
import ast
def get_wordlist_by_cat(cp, year, col):
    wordmap = {}
    cp_year = cp[cp.date.dt.year == year]
    for cat in np.unique(cp_year.catalogue):
        c_year = cp_year[cp_year['catalogue']==cat]
        wordlist = []
        for a in c_year[col]:
#            print(a)
            if a:
                words = a.replace('[','').replace(']','').replace(' ','').replace("'",'').split(',')
                for w in words:
                    wordlist.append(w)
#        print(wordlist)
        wordmap[cat] = np.unique(wordlist)
    return wordmap
```
